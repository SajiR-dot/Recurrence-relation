```python
'''T(n) = T(n-1) + c

T(n) = 1 if n <= 1 Base case condition
      T(n-1) if n>1

T(n) = T(n-1) + c 

lets substitute n-1 in the place of n 

T(n-1) = T(n-2) + c 

T(n) = T(n-2)+c + c
     = T(n-2) + 2c
 
Now lets us substitute n-2 in the place of n


T(n-2) = T(n-3)+c


T(n) = T(n-3)+ 3c

If we do the operation for k times

T(n) = T(n-k) + k.c

n-k = 1  
K= n-1

substituting for k

T(n) = T(1) + n-1.c
     = n


So the time complexity is O(n)

'''
```
